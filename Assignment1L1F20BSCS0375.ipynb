```python
import pandas as pd
import numpy as np
df=pd.read_csv('jsrt_metadata.csv')
df
```

```python
#Handling All missing Values
df['subtlety']=df['subtlety'].replace(np.nan,df['subtlety'].astype('float').mean())
```

```python
df['size']=df['size'].replace(np.nan, df['size'].astype('float').mean())
df
```

```python
df.isnull().sum()
```

```python
df['x']=df['x'].replace(np.nan, df['x'].astype('float').mean())
df['y']=df['y'].replace(np.nan, df['y'].astype('float').mean())
df['state']=df['state'].replace('non-nodule',np.nan)
df['state']=df['state'].replace(np.nan, df['state'].value_counts().idxmax())
df['position']=df['position'].replace(np.nan, df['position'].value_counts().idxmax())
df['diagnosis']=df['diagnosis'].replace(np.nan, df['diagnosis'].value_counts().idxmax())
df
```

```python
#Find the total number of patients of each disease (diagnosis) and plot in histogram.
import matplotlib.pyplot as plt
plt.hist(df['diagnosis'])
```

```python
#Q2
diseases=df['diagnosis']
diseases=list(set(diseases))# Storing All diseases in a list Without Duplication
fig, axes = plt.subplots(8, 4, figsize=(20, 15))#8 rows X 4 Columns= 32 Diseases
kk=0
for i in diseases:
    mylst=[0]*2
    itr=0
    for j in df['diagnosis']:
        if i==j:
            if df['state'][itr]=='malignant':
                mylst[0]+=1
            elif df['state'][itr]=='benign':
                mylst[1]+=1
        itr+=1
    ax = axes[kk//4, kk%4]
    kk+=1
    ax.pie(mylst,autopct='%1.0f%%',labels=['Malignant','Benign'],explode=(0,0.4))
    ax.set_title(i)
```

```python
#Q3
itr=0   
fig, axes = plt.subplots(8, 4, figsize=(20, 15))#8 rows X 4 Columns= 32 Diseases
kk=0
for i in diseases:
    mylst=[0]*2
    itr=0
    for j in df['diagnosis']:
        if i==j:
            if df['gender'][itr]=='Male':
                mylst[0]+=1
            elif df['gender'][itr]=='Female':
                mylst[1]+=1
        itr+=1
    ax = axes[kk//4, kk%4]
    kk+=1
    ax.pie(mylst,autopct='%1.0f%%', labels=['Male','Female'], explode=(0,0.2))
    ax.set_title(i)
```
